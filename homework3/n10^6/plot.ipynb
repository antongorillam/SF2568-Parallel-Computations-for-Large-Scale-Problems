```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("run_time.csv")
df
```

```python
plt.figure(figsize=(8,6), dpi= 100, facecolor='w', edgecolor='k')
plt.grid()
plt.title("Run time vs number of processors")
sns.lineplot(data=df, x="p", y="run_time", marker='o')
plt.savefig("run_time_vs_number_of_processors.png")
```
